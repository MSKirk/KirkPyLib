# Color Map Project

Comparing color maps

```python
import numpy as np           
import matplotlib.pyplot as plt
import os
import fnmatch
import matplotlib
import sunpy.map as map
import sunpy.cm as cm
from __future__ import division, print_function, absolute_import
import json
%matplotlib inline
```

load in AIA 304 data and select a sub section

```python
aia=map.Map('/Users/mskirk/data/AIA/20100726/All/aia_lev1_304a_2010_07_26t15_06_02_13z_image_lev1_fits.fits')
```

```python
test_data=np.log(aia.data[2001:3000,2001:3000])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(test_data, interpolation='none', cmap=cm.get_cmap('sdoaia304'))
plt.gca().invert_yaxis()

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(test_data, interpolation='none', cmap=cm.get_cmap('sohoeit304'))
plt.gca().invert_yaxis()
plt.show()
```

Load in custom color map

```python
path = '/Users/mskirk/py/lib/ColorMap/newcm2.jscm'

with open(path) as f:
    data = json.loads(f.read())
    name = data["name"]
    colors = data["colors"]
    colors = [colors[i:i + 6] for i in range(0, len(colors), 6)]
    colors = [[int(c[2 * i:2 * i + 2], 16) / 255 for i in range(3)] for c in colors]
custom_cmap = matplotlib.colors.ListedColormap(colors, name)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.imshow(test_data, interpolation='none', cmap=custom_cmap)
plt.gca().invert_yaxis()
plt.show()
```

All together now

```python
fig, axs = plt.subplots(1, 3, sharex=False, sharey=True, figsize=(20,6))

axs[0].imshow(test_data,interpolation='none', cmap=cm.get_cmap('sohoeit304'))
axs[0].axis('off')
axs[0].set_title("ColorTable = %s" % 'EIT 304')
axs[1].imshow(test_data,interpolation='none', cmap=cm.get_cmap('sdoaia304'))
axs[1].axis('off')
axs[1].set_title("ColorTable = %s" % 'AIA 304')
axs[2].imshow(test_data,interpolation='none', cmap=custom_cmap)
axs[2].axis('off')
axs[2].set_title("ColorTable = %s" % 'Custom')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('/Users/mskirk/py/lib/ColorMap/CustomCompare.eps',bbox_inches='tight', transparent=True)
plt.show()
```
